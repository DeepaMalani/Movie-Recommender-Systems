### Recommending Similar Movies

Create a matrix that has the user ids on one access and the movie title on another axis.

```python
movie_mat = df.pivot_table(index='user_id',columns='title',values = 'rating')

movie_mat.head()
```

```python
# Most rated movies
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings = movie_mat['Star Wars (1977)']
liarliar_user_ratings = movie_mat['Liar Liar (1997)']
starwars_user_ratings.head()
```

We can then use corrwith() method to get correlations between two pandas series:

```python
similar_to_starwars = movie_mat.corrwith(starwars_user_ratings)
similar_to_liarliar = movie_mat.corrwith(liarliar_user_ratings)

similar_to_starwars
```

Clean starwars_user_ratings by removing NaN values and using a DataFrame instead of a series:

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending = False).head(10)
```

Filter out movies that have less than 100 reviews 

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>50].sort_values('Correlation',ascending=False).head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
